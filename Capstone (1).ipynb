```python
import pandas as pd

df = pd.read_csv("D:/capstone project/zip/youtube_shorts_tiktok_trends_2025.csv")
df
```

```python
if "trend_label" in df.columns:
    trend_map={'rising':3, 'declining':0, 'seasonal':2, 'stable':1}
    df["trend_label"]=df["trend_label"].map(trend_map).fillna(0)
```

```python
df["trend_label"]
```

```python
if "platform" in df.columns:
    platform_map={'TikTok':1, 'YouTube':0}
    df["platform"]=df["platform"].map(platform_map).fillna(0)
```

```python
if "category" in df.columns:
    category_map={'Gaming':0, 'Food':1, 'Art':2, 'News':3, 'Beauty':4, 'Fitness':5, 'Pets':6,
       'Comedy':7, 'Science':8, 'DIY':9, 'Travel':10, 'Fashion':11, 'Lifestyle':12,
       'Tech':13, 'Music':14, 'Sports':15, 'Education':16, 'Finance':17, 'Automotive':18}
    df["category"]=df["category"].map(category_map).fillna(0)
```

```python
features = [
    "platform","category","duration_sec","completion_rate",
    "engagement_velocity","upload_hour","has_emoji"
]

X = df[features].copy()
y = df["trend_label"]
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test=train_test_split(
    X,y,test_size=0.2,stratify=y,random_state=42
)
```

```python
from sklearn.linear_model import LogisticRegression

model=LogisticRegression(max_iter=1000)
model.fit(X_train,y_train)
```

```python
y_pred=model.predict(X_test)
y_proba=model.predict_proba(X_test)[:,1]
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, classification_report, confusion_matrix

print("Accuracy :", accuracy_score(y_test, y_pred))
print("Precision :", precision_score(y_test, y_pred, average='weighted'))
print("Recall :", recall_score(y_test, y_pred, average='weighted'))
print("F1 Score :", f1_score(y_test, y_pred, average='weighted'))
```

```python
from sklearn.metrics import confusion_matrix

cm = confusion_matrix(y_test, y_pred)
print("\nConfusion Matrix:\n", cm)
```
